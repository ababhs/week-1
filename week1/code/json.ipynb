### 列举list元素

```python
x = [1, 2, 3, 4, 5]
for item in x:
    print(item)
for index, item in enumerate(x):
    print(index, item)
for index in range(len(x)):
    print(x[index])
```

### 列举dict元素

```python
x = {'a':1, 'b':2, 'c':3, 'd':4, 'e':5}
for item in x:
    print(item)
for key in x.keys():
    print(key, x[key])
for value in x.values():
    print(value)
for key, value in x.items():
    print(key, value)
```
